```python
from bs4 import BeautifulSoup
import requests
```

```python
# url="https://www.dinotube.com/category/cheating"
url="https://xhamster18.desi/pornstars"
```

```python
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.text,'html.parser')
```

```python
tags=soup.find_all('div',{"class":"pornstar-thumb-container"})
```

```python
for tag in tags:
    getter=tag.find('a',{"data-role":"pornstar-link"})
    name=getter.text if getter else "N/A"
    link_of_page=getter.get('href') if getter else "N/A"
    ranking_getter=tag.find('span',{"class":"pornstar-thumb-container__place exclude-tablet"})
    ranking=ranking_getter.text if ranking_getter else "N/A"
    img_getter=tag.find('img',{"class":"pornstar-thumb-container__original-image"})
    img_link=img_getter.get('src') if img_getter else "N/A"
    print(ranking.strip()," ",img_getter.get('alt'),"\nLink :",link_of_page)
#     print("Img : ",img_link)
    print('\n\n')
    #Entering into each page 
    url2=link_of_page
    response2=requests.get(url2)
    soup2=BeautifulSoup(response2.text,'html.parser')
    tags2=soup2.find_all("a",{"class","video-thumb__image-container role-pop thumb-image-container"})
    count=1;
    for tag2 in tags2:
        image_getter=tag2.find('img',{"class":"thumb-image-container__image"})
        print("\t",count,".",image_getter.get('alt'),"\n\tLink :",tag2.get('href'))
#         print("\t",image_getter.get('src').strip())
#         print("\n")
        count=count+1
    print("\n")
    print("**********************",count-1,"************************")
    print("\n")
```
